```python
# Шаг 1: Загружаем JSON файл
import json

# Укажите путь к вашему файлу
file_path = 'data/data_oos_plus.json'  # или data_small.json, data_imbalanced.json

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Шаг 2: Проверяем структуру данных
print(data.keys())
```

```python
oos_train_phrases = []

if 'oos_train' in data:
    for item in data['oos_train']:
        # item это список: [фраза, метка]
        if isinstance(item, list) and len(item) >= 1:
            oos_train_phrases.append(item[0])  # Берём только фразу

print(f"Найдено OOS фраз в train: {len(oos_train_phrases)}")
for i, phrase in enumerate(oos_train_phrases[:5]):
    print(f"{i+1}. {phrase}")
```

```python
# Шаг 4: Сохраняем в TXT файл (построчно)

output_file = 'data/clinc_ood_train.txt'

with open(output_file, 'w', encoding='utf-8') as f:
    for phrase in oos_train_phrases:
        f.write(phrase + '\n')
```
